```python
import keras
from keras import layers
(x_train , y_train) , (x_test , y_test) = keras.datasets.mnist.load_data()

model_mlp = keras.models.Sequential([
    keras.layers.Flatten( input_shape = [28,28]),
    keras.layers.Dense(200, activation = 'relu' ),
    keras.layers.Dense(250, activation = 'relu' ),
    keras.layers.Dense(100, activation = 'softmax' ),

])

model_mlp.compile( loss = 'sparse_categorical_crossentropy' , optimizer= 'adam' , metrics = ['accuracy'])
model_mlp.fit(x_train , y_train , epochs = 20 , validation_split=0.15)
```

```python
model_mlp.evaluate(x_test , y_test)
```

```python
model_conv = keras.models.Sequential([

    keras.layers.Conv2D(10 , 6 , activation= 'relu' , padding= 'same' , input_shape =[28,28,1]),
    keras.layers.AveragePooling2D(2),
    keras.layers.Conv2D(15 , 4 , activation= 'relu'),
    keras.layers.AveragePooling2D(2),
    keras.layers.Conv2D(20 , 3 , activation= 'relu'),
    keras.layers.AveragePooling2D(2),
    keras.layers.Flatten(),
    keras.layers.Dense(100, activation= 'elu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(10 , activation = 'softmax')
])
model_conv.compile( loss = 'sparse_categorical_crossentropy' , optimizer= 'adam' , metrics = ['accuracy'])
model_conv.fit(x_train , y_train , epochs = 20 , validation_split=0.15)
```

```python
model_conv.evaluate(x_test , y_test)
```
